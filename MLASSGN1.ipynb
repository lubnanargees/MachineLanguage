```python
# Import libraries
import numpy as np
from scipy import stats

# Copy to work on
df_clean = df.copy()

# a) Mean and Standard Deviation Method
mean_price = df_clean['price_per_sqft'].mean()
std_price = df_clean['price_per_sqft'].std()
upper_limit_a = mean_price + 3 * std_price
lower_limit_a = mean_price - 3 * std_price

df_trimmed_a = df_clean[(df_clean['price_per_sqft'] >= lower_limit_a) & 
                        (df_clean['price_per_sqft'] <= upper_limit_a)]

print("Mean & Std Method:", len(df_trimmed_a), "rows retained")

```

```python
# b) Percentile Method
lower_percentile = df_clean['price_per_sqft'].quantile(0.05)
upper_percentile = df_clean['price_per_sqft'].quantile(0.95)

df_trimmed_b = df_clean[(df_clean['price_per_sqft'] >= lower_percentile) & 
                        (df_clean['price_per_sqft'] <= upper_percentile)]

print("Percentile Method (1%-99%):", len(df_trimmed_b), "rows retained")
```

```python
# c) IQR Method
Q1 = df_clean['price_per_sqft'].quantile(0.25)
Q3 = df_clean['price_per_sqft'].quantile(0.75)
IQR = Q3 - Q1
lower_bound_c = Q1 - 1.5 * IQR
upper_bound_c = Q3 + 1.5 * IQR

df_trimmed_c = df_clean[(df_clean['price_per_sqft'] >= lower_bound_c) & 
                        (df_clean['price_per_sqft'] <= upper_bound_c)]

print("IQR Method:", len(df_trimmed_c), "rows retained")
```

```python
# d) Z-Score Method
z_scores = stats.zscore(df_clean['price_per_sqft'])
abs_z_scores = np.abs(z_scores)

df_trimmed_d = df_clean[abs_z_scores < 3]

print("Z-Score Method:", len(df_trimmed_d), "rows retained")
```

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Create subplots for each method
plt.figure(figsize=(15, 6))

# Original data
plt.subplot(1, 5, 1)
sns.boxplot(y=df_clean['price_per_sqft'])
plt.title("Original")
```

```python
# Mean & Std
plt.subplot(1, 5, 2)
sns.boxplot(y=df_trimmed_a['price_per_sqft'])
plt.title("Mean & Std")
```

```python
# Percentile
plt.subplot(1, 5, 3)
sns.boxplot(y=df_trimmed_b['price_per_sqft'])
plt.title("Percentile")
```

```python
# IQR
plt.subplot(1, 5, 4)
sns.boxplot(y=df_trimmed_c['price_per_sqft'])
plt.title("IQR")
```

```python
# Z-Score
plt.subplot(1, 5, 5)
sns.boxplot(y=df_trimmed_d['price_per_sqft'])
plt.title("Z-Score")

plt.tight_layout()
plt.show()
```

### ✅ Interpretation from the Box Plot (Q3)

The **box plot** visually shows how data is spread and where outliers lie — especially for `price_per_sqft`.

Here's how to interpret each one:

---

#### 📦 1. **Original Data**
- **Many outliers** far above the upper whisker.
- Indicates that raw data has extreme values — possibly incorrect or rare luxury properties.

#### 📊 2. **Mean & Std Deviation**
- Removes **some outliers**, but still shows a few extreme values.
- **Not great for skewed data**, as mean is sensitive to outliers.

#### 🎯 3. **Percentile Method (1%–99%)**
- Cuts off the **most extreme 1%** from both ends.
- Good balance — retains majority while removing extreme cases.
- Still some slight stretching at the top.

#### 📉 4. **IQR Method**
- Most **tight and clean plot**.
- Removes **almost all outliers** while preserving the data distribution.
- Best for **skewed distributions**, like real estate prices.

#### ⚖️ 5. **Z-Score**
- Similar to Mean/Std — works if data is **normally distributed**.
- Leaves outliers if data is skewed (like here).
- Not ideal for this case.

---

### 🏁 Conclusion:
> **IQR Method** is the most effective here.  
It removes outliers **without being too aggressive**, and the resulting box plot is clean and well-balanced.


```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Select the column
col = df['price_per_sqft']
```

```python
# Plot histogram
plt.figure(figsize=(10, 4))
sns.histplot(col, bins=50, kde=True)
plt.title("Histogram of price_per_sqft (Before Transformation)")
plt.xlabel("price_per_sqft")
plt.show()
```

```python
# BEFORE TRANSFORMATION
# -----------------------
print("Before Transformation:")
print("Skewness:", skew(col))
print("Kurtosis:", kurtosis(col))
```

```python
# Apply Log Transformation
# -----------------------
# Add 1 to avoid log(0)
log_col = np.log1p(col)
```

```python
# AFTER TRANSFORMATION
# -----------------------
print("\nAfter Log Transformation:")
print("Skewness:", skew(log_col))
print("Kurtosis:", kurtosis(log_col))
```

```python
# Plot histogram
plt.figure(figsize=(10, 4))
sns.histplot(log_col, bins=50, kde=True, color='green')
plt.title("Histogram of price_per_sqft (After Log Transformation)")
plt.xlabel("log(price_per_sqft + 1)")
plt.show()
```

The original `price_per_sqft` data was highly right-skewed with high kurtosis, indicating outliers and non-normality. A log transformation was applied to normalize the distribution. After transformation, skewness and kurtosis decreased significantly, and the histogram became more symmetric, improving the data’s suitability for statistical modeling and analysis.
